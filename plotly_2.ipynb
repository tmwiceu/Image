```python
import plotly.express as px

df = px.data.gapminder()
print(df.head())
df = px.data.gapminder().query("country=='Thailand'")
fig = px.line(df, x="year", y="lifeExp", title='Life expectancy in Thailand')
fig.show()
```

```python
import plotly.express as px

df = px.data.gapminder().query("continent=='Asia'")
print(df.head())
fig = px.line(df, x="year", y="gdpPercap",color="country",markers="True",symbol='country')

fig.show()
```
